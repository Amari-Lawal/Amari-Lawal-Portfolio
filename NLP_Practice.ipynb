```python
# Get all sentences from the dataframe 
# Get tokens of words in each sentence
# Add to the vocab
# Then count frequency of words between all sentences
 
def tokenizer_extraction(dataset):
  sent_doc = [sent for sent in dataset.iloc[:,1]]
  # Imports Counter Function
  from collections import Counter
  
  # Creates set which you add the vocab to and makes sure that there are no duplicates
  vocab = set() # However this means the vocab won't be in order and the indexes may change, INDEXES is the data format we will use to input into model/ 
  
  words_lis = [] 
  # Sets the first spam statement and setting it to spacy nlp object. 
  for sent1 in sent_doc:
    doc_sen = nlp(sent1)
    # Gets the tokens from the document getting rid of stop words, punctuation, digits and inculding lemmas
    words = [token.text for token in doc_sen if not token.is_stop and not token.is_punct and not token.is_digit and not token.is_currency and token.is_ascii] # May want to get rid of non unicode values with ord(token) < 128 on ascii table.
    for i in words:
      words_lis.append(i)
  
  # Adds the tokens in words to vocab set.
  for i in words_lis:
    vocab.add(i)

  
  # Counts the word frequency 
  word_freq = Counter(words_lis)
  common_words = word_freq.most_common(len(sent_doc))
  
  print(f'Vocab Size: {len(vocab)} ')
  print(f'Vocab Words: {vocab}')
  print("Word Frequency:",common_words)
  return vocab_lis, words_lis

vocab, words = tokenizer_extraction(df_spam) # If you look at the code in github the format will be abit of it is better in colab
```
